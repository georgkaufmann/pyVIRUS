# `pyVIRUS`

pyVIRUS, a program package for for modelling the spread and containment of a virus in a population.

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

import numpy as np
import os, csv, datetime
```

In this set of jupyter notebooks, we use data from the **Corona pandemic** to introduce
random-walk virus modelling and the description of a virus outbreak with differential equations.

----
## Read data set
As introductory example, we use the dataset from the **WHO** on infections in **Germany**
during the Corona pandemic, stored in the file `WHO_germany.csv` (see meta data).

We load this file with the `csv` library by:
- opening it with `open`
- reading it with `csv.reader`, and directly converting it to a `list` called `dataWHO`
- getting the number of lines (list elements) as `nlines`
- reformat the string data as
    - `times`: time array as datetime string(first element in `dataWHO` list, position 0)
    - `data`: infection data as numpy array (fifth and sixth element in `dataWHO` list, positions 4 and 5, new and accumulated cases)
    
Note the numerous conversion in this code snippet: `list`, `datetime` string, `int` ...

```python
infile = 'data/WHO_germany.csv'
f = open(infile,'r',newline='')
dataWHO = list(csv.reader(f, delimiter=';'))
f.close()
nlines = len(dataWHO)
times = []
data  = np.zeros([0,2],dtype='int')
print('=== initial shape of data array ===')
print(data.shape)

print('=== meta data lines ===')
for iline in range(nlines):
    if (dataWHO[iline][0][0] == '#'): 
        print(dataWHO[iline][0])
    else:
        #print(dataWHO[iline])
        y,m,d = dataWHO[iline][0].split('-')
        times.append(datetime.date(int(y),int(m),int(d)))
        data = np.append(data,[[int(dataWHO[iline][4]),int(dataWHO[iline][5])]],axis=0)

print('=== last data line ===')
print(times[-1],data[-1])
```

----
## Plot data set
We plot the Corona pandemic data for Germany, once as **current infections**,
once as **accumulated infections** ...

```python
fig,axs = plt.subplots(2,1,figsize=(10,6))
axs[0].set_title('Corona infections Germany (data: WHO)')
axs[0].set_ylabel('Current infections [Mill. people]')
axs[0].xaxis.set_major_formatter(mdates.DateFormatter("%Y-%m"))
axs[0].plot(times,data[:,0]/1e6)
axs[0].grid()

axs[1].set_ylabel('Total infections [Mill. people]')
axs[1].xaxis.set_major_formatter(mdates.DateFormatter("%Y-%m"))
axs[1].plot(times,data[:,1]/1e6)
axs[1].grid()

plt.tight_layout()
```

We can nicely observe the different infection waves, and in the accumulated cases the leveling off in early 2023.
